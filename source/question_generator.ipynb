```python
from pororo import Pororo
```

```python
qg = Pororo(task='qg', lang='ko')
```

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/pension/naver_news_all2.csv')
df
```

```python
line = df.iloc[0]['article']
line = line.strip()
line
```

```python
sub = df.iloc[0]['subject']
sub
```

```python
qg(answer=sub, context=line)
```

```python
df0 = df.copy()
df0 = df0.drop_duplicates(['subject', 'link'])
df0
```

```python
df0['question'] = ''
df0
```

```python
for idx in df0.index:
  line = df0['article'][idx].strip()
  sub = df0['subject'][idx]
  question = qg(answer=sub, context=line)
  df0['question'][idx] = question
  print(question)
```

```python
df0
```

```python
df1 = df0.copy()
df1
```

```python
for idx in df1.index:
  if ('@' in df1['question'][idx]) or ('http' in df1['question'][idx]):
    df1.drop(idx, axis=0, inplace=True)

df1
```

```python
df1.to_csv('news_question.csv', index=False)
```
