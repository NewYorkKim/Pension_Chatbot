```python
import pandas as pd
import json
import requests
```

```python
url_s = 'https://100lifeplan.fss.or.kr/openapi/api/psCorpList.json?key=63ce3944f4cf8cc6b9f861d17c72edf9&year=2021&quarter=4&areaCode=3'
res = requests.get(url_s)
contents_s = json.loads(res.text)

contents_s
```

```python
url_b = 'https://100lifeplan.fss.or.kr/openapi/api/psCorpList.json?key=63ce3944f4cf8cc6b9f861d17c72edf9&year=2021&quarter=4&areaCode=1'
res = requests.get(url_b)
contents_b = json.loads(res.text)

contents_b
```

```python
contents_s['list'][0]
```

```python
company_s = []
reserve_s = []
earnRate_s = []

for i in range(len(contents_s['list'])):
    company_s.append(contents_s['list'][i]['company'])
    reserve_s.append(contents_s['list'][i]['reserve'])
    earnRate_s.append(contents_s['list'][i]['earnRate'])
```

```python
company_b = []
reserve_b = []
earnRate_b = []

for i in range(len(contents_b['list'])):
    company_b.append(contents_b['list'][i]['company'])
    reserve_b.append(contents_b['list'][i]['reserve'])
    earnRate_b.append(contents_b['list'][i]['earnRate'])
```

```python
df_s = pd.DataFrame({'company': company_s, 'reserve': reserve_s, 'earnRate': earnRate_s})
df_b = pd.DataFrame({'company': company_b, 'reserve': reserve_b, 'earnRate': earnRate_b})
```
