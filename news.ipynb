```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# 한경

page = 1

url = f'https://search.hankyung.com/apps.frm/search.news?query=%EA%B0%9C%EC%9D%B8%EC%97%B0%EA%B8%88&sort=RANK%2FDESC%2CDATE%2FDESC&period=YEAR&area=ALL&mediaid_clust=HKPAPER%2CHKCOM&exact=&include=&except=&page={page}'

res = requests.get(url)
res.raise_for_status()

soup = BeautifulSoup(res.text, 'lxml')

ul = soup.find('ul', {'class': 'article'})
links = ul.find_all('a')

for link in links:
    url_a = link['href']
    if 'tag' not in link['href']:
        url_a = link['href']
        res_a = requests.get(url_a)
        soup_a = BeautifulSoup(res_a.text, 'lxml')
        text = soup_a.find('div', {'id': 'articletxt'})
        print(text)
```

```python
# 네이버

title_list = []
article_list = []
link_list = []

subjects = ['국민연금', '개인연금', '연금저축', 'irp개인연금', '연금보험']


for subject in subjects:
    for i in range(0, 10):
        start = 1 + (i*10)
    
        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/98.0.4758.102"}
        url = f'https://search.naver.com/search.naver?where=news&sm=tab_pge&query={subject}&sort=0&photo=0&field=0&pd=5&ds=2021.05.03&de=2022.05.03&cluster_rank=51&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:1y,a:all&start={start}'

        res = requests.get(url)
        res.raise_for_status()

        soup = BeautifulSoup(res.text, 'lxml')

        ul = soup.find('ul', {'class': 'list_news'})
        links = ul.find_all('a', {'class': 'info'})

        for link in links:
            if '네이버' in link.text:
                url_a = link['href']
                try:
                    res_a = requests.get(url_a, headers=headers)
                    soup_a = BeautifulSoup(res_a.text, 'lxml')
                    title = soup_a.find('h2', {'class': 'media_end_head_headline'})
                    article = soup_a.find('div', {'class': 'newsct_article'})
                    title_list.append(title.text)
                    article_list.append(article.text.replace('\n', ' ').replace('\t', ' ').replace('\xa0', ' '))
                    link_list.append(url_a)
                except:
                    pass
                
    print(f'{subject} is done!')
```

```python
result = pd.DataFrame({'title': title_list, 'article': article_list, 'link': link_list})
result
```

```python
result.to_csv('naver_news.csv', index=False)
```
